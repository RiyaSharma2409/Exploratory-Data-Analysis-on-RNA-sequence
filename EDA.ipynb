# **1. Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **2. Reading the dataset.**

```python
df = pd.read_excel('/content/HRNACountsDeseq2log2_data_18sub.xlsx')
```

# **3. Performing EDA**

**1. Dataset Summary**

```python
print("First 5 rows of the dataset:")
print(df.head())
```

```python
df.rename(columns={'Unnamed: 0': 'Gene Names'}, inplace=True)
df['Gene Names']
```

```python
print("\nDataset Info:")
print(df.info())
```

```python
print("\nSummary Statistics:")
print(df.describe(include='all'))
```

**2. Checking missing values**

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

**3. Univariate Analysis**

```python
df.hist(bins=30, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

**4. Bivariate Analysis**

```python
sns.scatterplot(data=df, x='G002-D0', y='Gene Names')
plt.title("Scatter plot of G002-D0 vs Gene Names")
plt.show()
```

**5. Correlation Analysis**

```python
# Compute correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

 **6. Outlier Detection**

```python
numeric_cols = df.select_dtypes(include='number').columns

plt.figure(figsize=(15, len(numeric_cols) * 3))

for i, col in enumerate(numeric_cols):
    plt.subplot(len(numeric_cols), 1, i+1)
    sns.boxplot(x=df[col], color='skyblue')
    plt.title(f'Boxplot for {col}')
    plt.tight_layout()

plt.show()
```
